```python
import numpy as np

data_train = np.genfromtxt('data_q0_train.csv', delimiter=',')
x_train = data_train[:,0].reshape(-1, 1)
y_train = data_train[:,1].reshape(-1, 1)

data_test = np.genfromtxt('data_q0_test.csv', delimiter=',')
x_test = data_test[:,0].reshape(-1, 1)
y_test = data_test[:,1].reshape(-1, 1)
```

We want to build a model that can predict y for unknown inputs x.

(a) (5 pts) Fit a linear model to the training data, and report mean squared error on the test data. Plot the data, fitted model, and predictions, clearly denoting the training, testing, and predicted points.

(b) (10 pts) Fit polynomial models of degrees 2, 3 and 4 to the training data, and report mean squared error (on both train and test sets) for all the models. Plot the data, the fitted models, and the predicted outputs.

```python
# (a)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
import matplotlib.pyplot as plt
reg = LinearRegression().fit(x_train, y_train)
reg_pred_test = reg.predict(x_test)
print("MSE:",mean_squared_error(reg_pred_test, y_test))
p_reg_test = plt.scatter(x_test, reg_pred_test, color='b', marker='o')
p_test = plt.scatter(x_test, y_test, color='g', marker='o')
x = np.arange(np.min(x_train[:,0])-5, np.max(x_train[:,0])+5, 0.02).reshape((-1,1))
reg_pred_train = reg.predict(x_train) 
p_reg_train = plt.scatter(x_train, reg_pred_train, color='r', marker='o')
p_train = plt.scatter(x_train, y_train, color='c', marker='o')

y = reg.predict(x)
plt.plot(x, y, c='blue', linewidth=0.5)
plt.legend((p_train, p_test, p_reg_train, p_reg_test),['train data', 'test data', 'train predictions', 'test predictions'])
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# (b)
poly2 = preprocessing.PolynomialFeatures(2)
x_train2 = poly2.fit_transform(x_train)
reg2 = LinearRegression().fit(x_train2, y_train)
x_test2 = poly2.fit_transform(x_test)
reg_pred_train2 = reg2.predict(x_train2)
reg_pred_test2 = reg2.predict(x_test2)
p_train = plt.scatter(x_train, y_train, color='c', marker='o')
p_test = plt.scatter(x_test, y_test, color='g', marker='o')
p_reg_train2 = plt.scatter(x_train, reg_pred_train2, color='r', marker='o')
p_reg_test2 = plt.scatter(x_test, reg_pred_test2, color='b', marker='o')
x2 = poly2.fit_transform(x)
y2 = reg2.predict(x2)
plt.plot(x, y2, c='blue', linewidth=0.5)
plt.legend((p_train, p_test, p_reg_train2, p_reg_test2),['train data', 'test data', 'train predictions', 'test predictions'])
plt.xlabel("x")
plt.ylabel("y")
plt.title("degree 2")
print("MSE(train):",mean_squared_error(reg_pred_train2, y_train))
print("MSE(test):",mean_squared_error(reg_pred_test2, y_test))
plt.show()
print("\n")
poly3 = preprocessing.PolynomialFeatures(3)
x_train3 = poly3.fit_transform(x_train)
reg3 = LinearRegression().fit(x_train3, y_train)
x_test3 = poly3.fit_transform(x_test)
reg_pred_train3 = reg3.predict(x_train3)
reg_pred_test3 = reg3.predict(x_test3)
p_train = plt.scatter(x_train, y_train, color='c', marker='o')
p_test = plt.scatter(x_test, y_test, color='g', marker='o')
p_reg_train3 = plt.scatter(x_train, reg_pred_train3, color='r', marker='o')
p_reg_test3 = plt.scatter(x_test, reg_pred_test3, color='b', marker='o')
x3 = poly3.fit_transform(x)
y3 = reg3.predict(x3)
plt.plot(x, y3, c='blue', linewidth=0.5)
plt.legend((p_train, p_test, p_reg_train3, p_reg_test3),['train data', 'test data', 'train predictions', 'test predictions'])
plt.xlabel("x")
plt.ylabel("y")
plt.title("degree 3")
print("MSE(train):",mean_squared_error(reg_pred_train3, y_train))
print("MSE(test):",mean_squared_error(reg_pred_test3, y_test))
plt.show()
print("\n")
poly4 = preprocessing.PolynomialFeatures(4)
x_train4 = poly4.fit_transform(x_train)
reg4 = LinearRegression().fit(x_train4, y_train)
x_test4 = poly4.fit_transform(x_test)
reg_pred_train4 = reg4.predict(x_train4)
reg_pred_test4 = reg4.predict(x_test4)
p_train = plt.scatter(x_train, y_train, color='c', marker='o')
p_test = plt.scatter(x_test, y_test, color='g', marker='o')
p_reg_train4 = plt.scatter(x_train, reg_pred_train4, color='r', marker='o')
p_reg_test4 = plt.scatter(x_test, reg_pred_test4, color='b', marker='o')
x4 = poly4.fit_transform(x)
y4 = reg4.predict(x4)
plt.plot(x, y4, c='blue', linewidth=0.5)
plt.legend((p_train, p_test, p_reg_train4, p_reg_test4),['train data', 'test data', 'train predictions', 'test predictions'])
plt.xlabel("x")
plt.ylabel("y")
plt.title("degree 4")
print("MSE(train):",mean_squared_error(reg_pred_train4, y_train))
print("MSE(test):",mean_squared_error(reg_pred_test4, y_test))
print()
plt.show()
```

Consider the following loss function of the scalar w.
\begin{align*}
\begin{split}
L(w) = \frac{1}{2}w^{2}\\
\end{split}
\end{align*}

We would like to minimize this function using gradient descent. So we start with initial value w(0) = 1 and update the weight at time step t as 
\begin{align*}
\begin{split}
w(t+1) = w(t) - \epsilon \frac{\partial L}{\partial w} \vert_t\\
\end{split}
\end{align*}

Compute the weights after the first three steps, i.e., compute w(1), w(2) and w(3) for each of the following choices of epsilon.
\begin{equation*} 
\epsilon = 0.5\\
\epsilon = 1\\
\epsilon = 2\\
\end{equation*}
Based on your observations, comment on how you expect the model to converge (if at all) for the different epsilon choices.

$$L(w)=\frac{1}{2}w^2$$
$$\frac{\partial L}{\partial w}=w$$
$$w(t+1)=w(t)-\epsilon w(t)=(1-\epsilon)w(t)$$
When $\epsilon=0.5$:<br>
$$w(1)=0.5w(0)=0.5$$
$$w(2)=0.5w(1)=0.5^2=0.25$$
$$w(3)=0.5w(2)=0.5^3=0.125$$
When $\epsilon=1$:<br>
$$w(1)=w(2)=w(3)=0$$
When $\epsilon=2$:<br>
$$w(1)=-w(0)=-1$$
$$w(2)=-w(1)=1$$
$$w(3)=-w(2)=-1$$
According to the above results. We can see when $\epsilon=$1, the loss function converges immediately. When $\epsilon=2$, the loss function will oscillate and never converge. If $|1-\epsilon|<1$ and $1-\epsilon\neq0$ which means $w\in(0,1)\cap(1,2)$. Then the loss function will converge but slower than $\epsilon=1$. $\epsilon\in(0,1)$, the loss function will converge in one direction. $\epsilon\in(1,2)$, the loss function will oscillating converge. For $\epsilon$ in other range, the loss function will diverge.

Q: Derive the equations to update the values of w5 and w1 after one iteration of backpropagation.

Let's suppose that at the very beginning all weight has the form $wi(0)$. Then after the first time forward propogation, the net gives an output $o=\sigma(w5\sigma(w1i1+w3i2)+w6\sigma(w2i1+w4i2))$ ($\sigma(x)$ is logistic function and all weight functions here have the value at the first iteration). Then compared the expected output $f$, we will get the squared loss function $\delta=\frac{1}{2}(t-o)^2$ ($t$ is the target).<br>
Then apply backpropogation for one iteration, get<br>
$$\frac{\partial\delta}{\partial o}=-(t-o)=o-t$$
$$\frac{\partial o}{\partial pre\_o}=(1-o)\cdot o$$
$$pre\_o=w5(0)h1(0)+w6(0)h2(0)$$
$$\frac{\partial pre\_o}{\partial w5(0)}=h1(0)$$
$$\frac{\partial\delta}{\partial w5(0)}=h1(0)\cdot(1-o)\cdot o\cdot(o-t)$$
Let $div=\frac{\partial\delta}{\partial w5(0)}$, then we know:
$$w5(1)=w5(0)-l\cdot div=w5(0)+l(t-o)(1-o)o\cdot h1(0)$$
Use the same method, get:
$$\frac{\partial pre\_o}{\partial h1(0)}=w5(0)$$
$$\frac{\partial h1(0)}{\partial pre\_h1(0)}=(1-h1(0))h1(0)$$
$$\frac{\partial pre\_h1(0)}{\partial w1(0)}=i1$$
$$\frac{\partial\delta}{\partial w1(0)}=(o-t)(1-o)\cdot o\cdot i1\cdot w5(0)(1-h1(0))h1(0)$$
Then we get:
$$w1(1)=w1(0)+l(t-o)o(1-o)i1\cdot w5(0)(1-h1(0))h1(0)$$

Try to format similarly to scikit-learn's models. The template of the class is given. The `fit()` method must take as input `X, y` and a choice of `update_rule` as `'sgd'` or `'sgd_momentum'` or `'Adagrad'`. The `predict()` method takes an `X` value (optionally, an array of values). Use your new gradient descent regression to predict the data given in 'q5_samples.csv', for 15 epochs, using learning rates: `[0, .0001, .001, .01, 0.1, 1, 10, 100]` and regularization (ridge regression) constants: `[0, 10, 100]`. **(8 points)**

```python
%matplotlib inline
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

class LinearRegression:
    
    def __init__(self, learning_rate, regularization, n_epoch):
        self.learning_rate = learning_rate
        self.n_epoch = n_epoch
        self.regularization = regularization
        self.coef = np.zeros(5) #one coef for each of the terms
        self.eps = 10**-6 #used only for adagrad
        self.cache = np.zeros(self.coef.shape) #used only for adagrad
        self.mu = 0.9 #used in momentum
        self.momentum = np.zeros(self.coef.shape)
        self.mse = -1 #used to choose the best learning rate - regularization parameter pair
        
    def sgd(self, gradient):
        '''
        updates self.coef based on gradient using SGD
        '''
        self.coef = self.coef - self.learning_rate*gradient
        
    def adagrad(self, gradient):
        '''
        updates self.coef based on gradient using Adagrad
        '''
        self.cache += gradient**2
        G = np.diag(self.learning_rate/np.sqrt(self.cache + self.eps))
        self.coef = self.coef - G.dot(gradient)
    
    def sgd_momentum(self, gradient):
        '''
        updates self.coef based on gradient using Sgd_momentum
        '''
        self.momentum = self.mu*self.momentum + self.learning_rate*gradient
        self.coef = self.coef - self.momentum
    
        
    def fit(self, X, y, update_rule='sgd', plot=False):
        '''
        Fits the model for the given X, y. It uses the update rule mentioned by the parameter update_rule
        and displays a plot of the coefficients vs epochs and mse vs epochs if plot is True. 
        
        -> use get_features to get the features from X
        -> for epoch in epochs:
            iterate through all x, y.
                compute prediction using linearPredict.
                compute gradient.
                pass this gradient to the respective update rule and update the coefficients
                keep track of mse and coefficients
        -> plot if required
        
        '''
        mse_list = []
        w_list = []
        X = self.get_features(X)
        forward = self.linearPredict(X)
        for epoch in range(self.n_epoch):
            temp = np.random.choice(len(X),len(X),replace=False)
            for i in temp:
                gradient = -1*(y[i] - forward[i])*X[i] + self.regularization*self.coef
                if update_rule=='sgd':
                    self.sgd(gradient)
                if update_rule=='sgd_momentum':
                    self.sgd_momentum(gradient)
                if update_rule=='adagrad':
                    self.adagrad(gradient)
                forward = self.linearPredict(X)
            mse = mean_squared_error(y, forward)
            
            mse_list.append(mse)
            w_list.append(self.coef)
        self.mse = mse
        w_list = np.array(w_list)
        if plot:
            ax1 = plt.subplot(121)
            ax2 = plt.subplot(122)
            w0=ax2.plot(range(1,self.n_epoch+1),w_list[:,0])
            w1=ax2.plot(range(1,self.n_epoch+1),w_list[:,1])
            w2=ax2.plot(range(1,self.n_epoch+1),w_list[:,2])
            w3=ax2.plot(range(1,self.n_epoch+1),w_list[:,3])
            w4=ax2.plot(range(1,self.n_epoch+1),w_list[:,4])
            ax1.plot(range(1,self.n_epoch+1),mse_list)
            ax1.title.set_text('MSE')
            ax2.title.set_text(r'$\omega$')
            ax2.legend(['w0','w1','w2','w3','w4'])
            plt.tight_layout()
            plt.suptitle(update_rule)
            plt.show()


    def get_features(self, X):
        '''
        X (input) is the nx2 dimensional array - n data points with X_1 and X_2.
        It returns a nX5 dimensional array - n data points with 1, X_1, X_1X_2, X_1^2X_2^2, X_2^3.
        '''
        ones = np.ones(len(X))
        X_1 = X[:,0]
        X_2 = X[:,1]
        temp = np.c_[ones,X_1,X_1*X_2,(X_1**2)*(X_2**2),X_2**3]
        return temp
    
    def linearPredict(self, X_features):
        '''
        returns the dot product of X and self.coef
        '''
        return  X_features.dot(self.coef)
    
    def get_mse(self):
        return self.mse
```

$\textbf{2)}$<br>
It seems that adagrad algorithm allows larger initial learning rate. For sgd and sgd_momentum, if we set the learning rate too large. The loss function will oscillate around the local minimum and might never converge (which might lead to an infinite MSE value). In adagrad algorithm every time the learning rate is scaled such that the loss function will decrease more and more slowly. That is why adagrad tend to converge slowly. Because the loss function (MSE) will decrease more and more slowly with more iterations, MSE will not oscillate too much and adagrad always give a MSE that is not too large (overflow).
